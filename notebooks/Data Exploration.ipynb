```python
import os 
import glob
directory = "../data/robust_25_query"
query_files = directory + "/*.json" 
files = glob.glob(query_files)
files
```

```python
import ast
hard_query_list = ast.literal_eval(open(os.path.join(directory, "hard_query.txt")).readline())
hard_query_dict = {}
for hard_query in hard_query_list:
    hard_query_dict[hard_query[0]] = hard_query[1]
print(hard_query_dict)
    
import json 
for file in files: 
    query_id = file.split("/")[-1].split(".")[0]
    query_dict = json.load(open(file))
    print("{} {} {}".format(query_id, hard_query_dict[int(query_id)], query_dict["query_text"]))
```
